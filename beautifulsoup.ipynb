### Importing libraries needed

```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

### Base URL of the site to be analyzed

```python
SITE_URL = 'https://es.wikipedia.org'
```

### Creating a new chrome window 

```python
serv = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service = serv)
browser.get(SITE_URL)
```

### Fuction to rendering the browser

```python
def make_request(browser, relative_path):
    # Making the request and rendering the browser
    browser.get(SITE_URL + relative_path)
    
    # Simulating vertical scrolling for handling lazy load
    check_height = browser.execute_script('return document.body.scrollHeight;')
    while True:
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(3)
        height = browser.execute_script('return document.body.scrollHeight;')
        if height == check_height: 
            break 
        check_height = height
    
    # Getting HTML content and passing it to BeautifulSoup for scraping analysis
    return BeautifulSoup(browser.page_source, 'html.parser')
```

### Getting HTML content for president's table list

```python
soup = make_request(browser, '/wiki/Anexo:Presidentes_de_Colombia')
```

### Finding the section where tables are contained 

```python
content = soup.find(class_ = 'mw-parser-output')
```

```python
# Total tables
tables = content.find_all('table')
len(tables)
```

### Getting table rows HTML content

```python
tr_elements = content.find_all('tr')
```

### Saving columns HTML content in a list

```python
td_elements = []
for tr in tr_elements:
    td_elements.append(tr.find_all('td'))
```

### Saving the presidents information in a dictionary

```python
presidents = []
for col in td_elements:
    try:
        presidents.append({
            'full_name': col[3].getText(),
            'time_period': col[1].getText(),
            'politic_party': col[5].getText()
        })
    except:
        print(col)
```
